## Data Notes
The original CSV export did not include explicit latitude/longitude columns.
To preserve spatial accuracy, this project uses the GeoJSON version of the
dataset and GeoPandas to handle geometry directly.

## Visualization Approach
Because the dataset contains true GIS geometry, visualization is handled
using GeoPandas' built-in plotting rather than manual latitude/longitude
scatter plots.

## Observations
- Parks appear clustered in some areas more than others
- Central and river-adjacent areas show higher density
